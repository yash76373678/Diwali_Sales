```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
total_revenue=df['Amount'].sum()
print("Total sales revenue during Diwali:",total_revenue)
```

```python
df['Product_Category'].unique()
```

```python
top_products = df.groupby('Product_Category')['Orders'].sum().sort_values(ascending=False)
print('Top-Orders products during Diwali:\n', top_products)
```

```python
grouped_data = df.groupby('Product_Category')['Orders'].sum().sort_values(ascending=False)
```

```python
plt.bar(grouped_data.index, grouped_data.values)
plt.xlabel('Product Category')
plt.ylabel('Orders')
plt.title('Orders of different product categories during Diwali')
plt.xticks(rotation=90)
plt.show()
```

```python
top_sales=df.groupby('Product_Category')['Amount'].sum().sort_values(ascending=False)
print("Top-sales products during Diwali:\n",top_sales)
```

```python
grouped_data=df.groupby('Product_Category')['Amount'].sum().sort_values(ascending=False)
```

```python
plt.bar(grouped_data.index,grouped_data.values)
plt.xlabel("Product Category")
plt.ylabel("Amount")
plt.title('Sales of different product categories during Diwali')
plt.xticks(rotation=90)
plt.show()
```

```python
grouped_df = df.groupby(['Age Group', 'Gender']).size().unstack()

x_labels = grouped_df.index
x = np.arange(len(x_labels))
width = 0.35

female_counts = grouped_df['F'].values
male_counts = grouped_df['M'].values

plt.figure(figsize=(10, 10))
#figure(figsize=(WIDTH_SIZE,HEIGHT_SIZE))
plt.bar(x - width/2, female_counts, width)
plt.bar(x + width/2, male_counts, width)

plt.xlabel("Age Group")
plt.ylabel("Count")
plt.title("Count of Male and Female in Each Age Group")
plt.xticks(x,x_labels)
plt.legend(['F','M'])

plt.show()
```

```python
grouped_data=df.groupby('Age Group')['Amount'].sum().sort_values(ascending=False)
plt.bar(grouped_data.index,grouped_data.values)
plt.xlabel("Age Group")
plt.ylabel("Total Amount")
plt.title('Total Amount by Age Group')
plt.show()
```
